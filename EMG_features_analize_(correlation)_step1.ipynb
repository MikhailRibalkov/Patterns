При классификации сигналов EMG для распознавания жестов, возникают ошибки, которые обусловлены изменением состояния человека во времени и соответственно изменением сигналов считываемых с поверхности кожи.
Предположим что сигнал имеет признаки, которые при изменении сигнала сохраняют автокорреляцию и позволяют произвести корректировки для лучшего распознавания.

```
import scipy.io
import csv
import numpy as np
import requests
import matplotlib.pyplot as plt


#загрузим файл с признаками (6 человек, 9 движений, по 35 признаков для каждого движения)
path_to_file = 'https://github.com/RF-Lab/emg_platform/raw/master/data/Future_all.mat'
!wget $path_to_file

mat = scipy.io.loadmat('Future_all.mat')
all = mat['P_all']

#выберем признаки для одного типа движения одного человека
men = all[0][0][0][0]
```

Построим графики признаков и графики показывающие степень автокорреляции для каждого признака

```
plt.plot(men[0])
plt.title('Integrated EMG (IEMG)')
plt.show()

cor = np.correlate(men[0], men[0], mode='same')
plt.plot(cor)
plt.title('correlate')
plt.show()
```
